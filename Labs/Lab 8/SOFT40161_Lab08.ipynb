## Example 1: Create a line plot.

Purpose: Ideal for tracking changes or trends across time intervals (e.g., days, months, years).
Key Elements:
- X-axis: Represents independent variable (e.g., time, dates, years).
- Y-axis: Represents the variable being measured.

Design Tips:
- Use markers for data points to highlight changes.
- Limit the number of lines to prevent clutter.
- Incorporate annotations for significant events or changes.

Customization:
- Style with colors and line types (e.g., dashed, solid).
- Add gridlines for better readability.


```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5, 8]
y = [2, 3, 5, 9, 10, 12]
z = [.5, 1, 3, 16, 20]

# Create the line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot Example")

# Show the plot
#plt.savefig('linePlot.png')
plt.show()
```

Explanation:

Purpose: This example demonstrates how to create a basic line plot using Matplotlib's plt.plot() function.
Key Components:
x and y: Define the data points.
plt.plot(): Plots x against y.
Customization: Added color (blue), markers (o), line style (-), and line width (2).
Annotations: Labels (xlabel, ylabel) and a title (title) improve the plots clarity and presentation.

## Example 2: Multiple Lineplot

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 9, 10]
z = [.5, 1, 3, 16, 20]

# Create the line plot
plt.plot(x, y, label = 'x data', color = 'green', marker = 'v', linestyle = ':', linewidth = 2)
plt.plot(x, z, label = 'z data', color='red', marker='o', linestyle='-', linewidth=3)

# Add labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot Example")
plt.grid(True, linestyle=':', linewidth=0.5)
plt.legend()

# Show the plot
#plt.savefig('linePlot2.png')
plt.show()
```

**Explanation:**

- Line Styles: - (solid), -- (dashed), and -. (dash-dot) provide visual differentiation.

**Customization:**
- title, xlabel, and ylabel enhance clarity.
- grid improves readability.
- legend helps identify each line.
- This approach effectively visualizes trends for multiple categories over time.

## Example 3: Subplot using Matplotlib

A detailed documentation of is here- <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html"> (Click to open)</a>

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 50)
y1 = x**2
y2 = np.sqrt(x)
y3 = np.log1p(x)  # log(1 + x)
y4 = np.exp(x / 10)
y5 = x**3
y6 = np.expm1(x)

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 8))

# First plot
axes[0, 0].plot(x, y1, label="x^2", color="blue")
axes[0, 0].set_title("Quadratic")
axes[0, 0].legend()

# Second plot
axes[0, 1].plot(x, y2, label="x", color="green")
axes[0, 1].set_title("Square Root")
axes[0, 1].legend()

# Third plot
axes[1, 0].plot(x, y3, label="log(1 + x)", color="red")
axes[1, 0].set_title("Logarithmic")
axes[1, 0].legend()

# Fourth plot
axes[1, 1].plot(x, y4, label="exp(x/10)", color="purple")
axes[1, 1].set_title("Exponential")
axes[1, 1].legend()

#Fifth plot
axes[2,0].plot(x,y5, label = 'x^3', color = "orange")
axes[2,0].set_title("Cubic")
axes[2,0].legend()

#Sixth plot
axes[2,1].plot(x,y6, label = 'exp(x)-1', color = "black")
axes[2,1].set_title("exp(x)-1")
axes[2,1].legend()

# Adjust layout
plt.tight_layout()
plt.savefig('Subplot1.png')
plt.show()
```

**Explanation:**
- Data Preparation: Each subplot uses a unique mathematical function for variety.
- Subplots Grid: plt.subplots(2, 2) creates a 2x2 grid for visualisations.
- Labels and Legends: Simple labels and titles make plots self-explanatory.
- Minimal Code: Achieves effective visualisation with minimal complexity.
- `plt.subplots(2, 2)`: Creates a 2x2 grid of subplots.
axs is a 2D array where each element represents a subplot (e.g., axs[0, 0] for the top-left subplot).
Plotting Data:
- Use `axs[row, col].plot()` to specify the subplot to plot data on. Customize each subplot with unique colors, markers, and line styles.
- This example demonstrates how to visualise multiple datasets compactly.

## Example 4: Subplot using Matpltlib-2. Use of grid, `xlim` and `ylim`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 2 * np.pi, 100)  # Values from 0 to 2

# Calculate trigonometric functions
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)

# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# First subplot: Sine
axes[0].plot(x, sin_y, color='blue', linestyle='-', linewidth=2)
axes[0].set_title("Sine Function")
axes[0].set_xlabel("X (radians)")
axes[0].set_ylabel("sin(X)")
axes[0].grid(True)

# Second subplot: Cosine
axes[1].plot(x, cos_y, color='green', linestyle='--', linewidth=2)
axes[1].set_title("Cosine Function")
axes[1].set_xlabel("X (radians)")
axes[1].set_ylabel("cos(X)")
axes[1].grid(True)

# Third subplot: Tangent
# Note: Limiting Y-axis to avoid extreme tangent values
axes[2].plot(x, tan_y, color='red', linestyle='-.', linewidth=2)
axes[2].set_title("Tangent Function")
axes[2].set_xlabel("X (radians)")
axes[2].set_ylabel("tan(X)")
axes[2].set_ylim(-10, 10)  # Limit y-axis
axes[2].grid(True)

#Custom x axis
ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
labels = [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']

for ax in axes: 
    ax.set_xticks(ticks)
    ax.set_xticklabels(labels)

# Adjust layout and show
plt.tight_layout()
plt.savefig('Subplot2.png')
plt.show()
```

Explanation:

Purpose: Visualizes sine, cosine, and tangent functions side-by-side.
Key Components:
np.linspace(0, 2 * np.pi, 100): Generates evenly spaced x-values over a full cycle.
Each subplot focuses on a specific trigonometric function.
set_ylim(-10, 10): Ensures tangent values are contained within viewable limits.
Advantages: Trigonometric plots are essential for understanding periodic behaviors.

## Example 5: Histogram and Boxplot
**Histogram:** A histogram visualizes the distribution of a continuous variable by dividing data into bins and displaying the frequency of observations in each bin. It often includes a density curve (KDE) to highlight the distribution pattern.

**Box Plot:** A box plot summarizes data distribution using the median, quartiles, and outliers. It shows variability and highlights differences across categories, making it useful for comparing multiple datasets or visualizing skewness and potential anomalies.


- seaborn: Libraries used for creating fancy ways of visualisations.
- np.random.seed(5): Sets a random seed to ensure reproducibility of random data.
A pandas.DataFrame is created with two columns:
- Category: Randomly chosen values from `["A", "B", "C"]` for 100 rows.
- Values: Randomly generated 100 numbers from a standard normal distribution.

**Setting Up the Plot:**
- `plt.figure(figsize=(12, 5))`: Creates a figure with a size of 12 inches by 5 inches.
- `plt.subplot(1, 2, 1)`: Specifies that the first subplot will be used (1 row, 2 columns, 1st plot).
- `sns.histplot(data["Values"], kde=True, color="green")`: Plots a histogram of the Values column.
- `plt.subplot(1, 2, 2)`: Specifies the second subplot will be used (1 row, 2 columns, 2nd plot).
- `sns.boxplot(x="Category", y="Values", data=data, palette="Set2")`: Creates a boxplot to compare the distribution of Values across Category groups `(A, B, C)`.
- Uses the `"Set2"` palette for colors.
- `plt.title("Boxplot by Category")`: Adds a title to the boxplot.
- `plt.tight_layout()`: Ensures that the subplots do not overlap.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
np.random.seed(5)
data = pd.DataFrame({
    "Category": np.random.choice(["A", "B", "C"], size=100),
    "Values": np.random.randn(100)
})

# Histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data["Values"], kde=True, color="green")
plt.title("Histogram with KDE")

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x="Category", y="Values", data=data,palette="Set2")
plt.title("Boxplot by Category")

plt.tight_layout()
plt.savefig("histBox.png")
plt.show()
```

## Example 6: Interactive Plot

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    "City": ["London", "Paris", "Berlin", "New York", "Tokyo"],
    "Population (M)": [9.0, 11.1, 3.8, 8.4, 14.0],
    "Continent": ["Europe", "Europe", "Europe", "North America", "Asia"]
}
df = pd.DataFrame(data)

# Create an interactive scatter plot
fig = px.scatter(
    df,
    x="City",
    y="Population (M)",
    color="Continent",
    size="Population (M)",
    title="City Populations by Continent",
)

fig.update_layout(
    title_font_size=18,
    hovermode="closest",
    showlegend=True,
    legend_title_text="Continent"
)

fig.show()
```

**Explanation**
- Plotly for Interactivity: `px.scatter` creates interactive scatter plots with tooltips.
- Built-in filters enable dynamic updates based on dropdown selections or hover interactions.
- Exploration Made Simple: Interactive plots make complex datasets user-friendly.
- Engagement: Allows end-users to focus on their areas of interest.


## Example 7: Multidimensional Scatter plot.

A multidimensional scatter plot visualises relationships between three or more variables. Typically, the x, y, and z axes represent three dimensions, while color, size, or shape of points encode additional variables. It's useful for analyzing complex, high-dimensional data relationships.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(42)
x = np.random.rand(50) * 100  # X values
y = np.random.rand(50) * 100  # Y values
sizes = np.random.rand(50) * 1000  # Bubble sizes
colors = np.random.rand(50)  # Color values

# Plotting
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)

# Adding labels and title
plt.title("Multi-dimensional Scatter Plot", fontsize=14)
plt.xlabel("X-axis: Feature 1", fontsize=12)
plt.ylabel("Y-axis: Feature 2", fontsize=12)

# Color bar for additional dimension
color_bar = plt.colorbar(scatter)
color_bar.set_label("Color Dimension")

# Display the plot
plt.savefig('MultScatter.png')
plt.show()
```

## Example 8: multidimensional heatmap

A multidimensional heatmap is a visual representation of data where values are encoded as colors in a grid. It can display relationships across multiple variables, often combining rows, columns, and annotations, making it effective for exploring patterns and correlations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random multi-dimensional data
np.random.seed(42)
data = np.random.rand(10, 12)  # 10 rows (e.g., samples) and 12 columns (e.g., features)

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt=".2f", cmap="YlGnBu", cbar=True)

# Add titles and labels
plt.title("Multi-Dimensional Heatmap", fontsize=16)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Samples", fontsize=12)

# Display the plot
plt.savefig('MultHeatMap.png')
plt.show()
```

**Explanation**
- Data: The 10x12 matrix represents 10 samples and 12 features.
- `annot=True`: Displays the actual values on the heatmap cells.
- Color Map (cmap): Enhances visualization by using a perceptually uniform palette.
- Grid: Makes relationships between features visually apparent.
- Color Bar: Adds another layer of information about data magnitude.
- Brighter or darker areas highlight significant values or trends. Quickly identify high or low feature values across samples.

**Saving multiple plots dring the exeution time**
Heres a small Python script that saves n number of plots separately using Seaborn's planet dataset. It generates a scatterplot for the first n combinations of variables in the dataset and saves them as separate image files.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os

# Load the Seaborn 'planet' dataset
penguins = sns.load_dataset("penguins")

# Specify the number of plots to save
n = 5

# Create 'multipleFigure' folder if it doesn't exist
folder_path = "multipleFigure"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Get the column names of the dataset
columns = penguins.columns

# Generate and save 'n' scatter plots
for i in range(n):
    # Create a plot for each pair of variables
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=penguins, x=columns[i % len(columns)], y=columns[(i + 1) % len(columns)],hue="species")
    plt.title(f"Scatter Plot of {columns[i % len(columns)]} vs {columns[(i + 1) % len(columns)]}")
    
    # Save the plot inside 'multipleFigure' folder
    plt.savefig(f"{folder_path}/plot_{i+1}.png")
    plt.close()

print(f"Saved {n} plots successfully inside '{folder_path}' folder.")
```

**Explanation:**
- Dataset: Loads the planet dataset using Seaborns load_dataset.
- Loop: The loop generates n scatter plots by iterating over pairs of columns in the dataset.
- Saving Plots: Each plot is saved with a unique filename, like plot_1.png, plot_2.png, etc.
- Folder Creation: `os.makedirs(folder_path)` checks if the `multipleFigure` folder exists, and if not, it creates it.
- Saving Plots: The `plt.savefig(f"{folder_path}/plot_{i+1}.png")` code saves each plot inside the multipleFigure folder with unique filenames like plot_1.png, plot_2.png, etc.
- This ensures that the plots are organized and saved inside the multipleFigure folder
- Closing Plots: `plt.close()` ensures that each plot is closed after saving, preventing overlap when creating multiple plots.
