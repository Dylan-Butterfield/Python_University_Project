## Example 01: Hello World program using Tkinter:

We are going to create our first GUI with the `Hellow NTU Students` message 

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Hello World App")  # Set the window title
root.geometry("500x250")       # Set the window size

# Create a label widget
label = tk.Label(root, text="Hello, NTU Students! How are you today?", font=("Helvetica", 16))
label.pack(pady=100)  # Add some padding and display the label

# Run the Tkinter event loop
root.mainloop()
```

**Explanation:**
- `tk.Tk()`: Creates the main application window.
- `title`: Sets the title of the window.
- `geometry`: Defines the dimensions of the window.
- `Label`: Adds a text label with customizable font and size.
- `pack`: Places the label in the center with padding.
- `mainloop`: Keeps the application running, waiting for user interaction.
- When you run this script, a small window will appear with "Hello, World!" displayed in the center.

## Example 02: A Tkinter Label in GUI

A Label in Tkinter is a widget used to display text, images, or other content in a GUI application. It is a simple and non-interactive element, often used to provide instructions, headings, or display output.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Label Example")
root.geometry("400x200")

# Create a Label widget
label = tk.Label(
    root,
    text="Welcome to Tkinter!",
    font=("Arial", 16),
    bg="lightblue",
    fg="darkblue",
    padx=20,
    pady=10,
    relief="raised"
)
label.pack(pady=50)  # Place the label in the window with vertical padding

# Run the application
root.mainloop()
```

**Output:**
A window titled "Label Example" with:
- A blue-colored text label that says "Welcome to Tkinter!".
> The label has a light blue background, dark blue text, padding, and a raised border style.
- You can customize the text, colors, and font to suit your needs!

## Example 03: Tkinter Button Widget

A Button in Tkinter is a widget that allows users to trigger an action or execute a command when clicked. It is one of the most commonly used widgets for interactive GUI applications. A button can use to perform specific tasks when clicked (e.g., submit, cancel, open files).
Can display text, images, or both.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Button Click Example")
root.geometry("300x200")

# Define the button click handler function
def on_button_click():
    label.config(text="You clicked the button!")  # Update the label text

# Create a Label widget
label = tk.Label(root, text="Click the button below", font=("Arial", 14))
label.pack(pady=20)

# Create a Button widget
button = tk.Button(
    root, 
    text="Click Me", 
    font=("Arial", 12), 
    bg="lightblue", 
    command=on_button_click
)
button.pack(pady=20)

# Run the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Button Click Example")
root.geometry("300x200")

# Define the button click handler function
def Close():
    root.destroy()  # Update the label text

# Create a Label widget
label = tk.Label(root, text="Click the button below", font=("Arial", 14))
label.pack(pady=20)

# Create a Button widget
button = tk.Button(
    root, 
    text="Click Me", 
    font=("Arial", 12), 
    bg="lightblue", 
    command=Close
)
button.pack(pady=20)

# Run the Tkinter event loop
root.mainloop()
```

**Explanation**

- `root`: The parent container (e.g., a Tk window or Frame).
- `text`: The text displayed on the button.
- `command`: A function to execute when the button is clicked.

options: Various attributes to style and configure the button (optional). Common Options for Button:
- `text`: Text displayed on the button.
- `command`: Function to call when the button is clicked.
- `font`: Font style and size, e.g., ("Helvetica", 14).
- `bg / fg`: Background and foreground (text) colors.
- `padx / pady`: Padding around the button content.
- `width / height`: Size of the button (in text units, not pixels).


## Example 04: Drawing Shapes on a Canvas

The Canvas widget in Tkinter is a versatile widget that allows you to create custom drawings, shapes, and designs. It can also be used to display images, add text, or embed other widgets.

```python
import tkinter as tk
import random as rd


# Create the main window
root = tk.Tk()
root.title("Canvas Example")
root.geometry("400x300")

# Create a Canvas widget
canvas = tk.Canvas(root, width=300, height=200, bg="lightgray")
canvas.pack(pady=20)

# Draw a rectangle
canvas.create_rectangle(50, 50, 150, 100, fill="blue", outline="black", width=2)

# Draw an oval (ellipse)
canvas.create_oval(170, 50, 270, 100, fill="red", outline="black", width=2)

# Draw a line
canvas.create_line(50, 150, 270, 150, fill="green", width=3)

# Draw text
canvas.create_text(150, 180, text="Tkinter Canvas", font=("Arial", 12), fill="black")

#Create Polygon
points = []

for x in range(10):
    rand_numb = rd.randint(50,100)
    points.append(rand_numb) 

canvas.create_polygon(points, outline = "blue", fill = "orange", width = 2)

# Run the application
root.mainloop()
```

**Output:**
A window with:
- A blue rectangle on the canvas.
- A red oval next to the rectangle.
- A green horizontal line at the bottom of the canvas.
- A text saying "Tkinter Canvas".

Common Methods for Canvas:
- `create_rectangle(x1, y1, x2, y2, options...)`: Draws a rectangle.
- `create_oval(x1, y1, x2, y2, options...)`: Draws an ellipse or circle.
- `create_line(x1, y1, x2, y2, options...)`: Draws a line.
- `create_text(x, y, text="...", options...)`: Displays text.
- `create_polygon(coordinates, options...)`: Draws a polygon.
This is a great widget for custom graphics, simple games, or data visualizations!

## Example 05: Layout Management Methods

In Tkinter, Layout Management Methods are used to control the placement and arrangement of widgets within a window or a container. These methods determine how widgets like buttons, labels, and entry fields are displayed in a graphical user interface (GUI).

### Example 05a: Layout Management Methods: pack()

Description:
- Automatically positions widgets in a block-like structure (top, bottom, left, right).
- Uses options like `side`, `fill`, and `expand` for customization.

Use Cases:
- Simple layouts with minimal customization.
- Quickly stacking widgets vertically or horizontally.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Pack Layout Example")
root.geometry("300x200")

# Add widgets using pack
tk.Label(root, text="Top", bg="lightblue").pack(side="top", fill="x")
tk.Label(root, text="Bottom", bg="lightgreen").pack(side="bottom", fill="x")
tk.Label(root, text="Left", bg="pink").pack(side="left", fill="y")
tk.Label(root, text="Right", bg="yellow").pack(side="right", fill="y")

# Run the application
root.mainloop()
```

**Explanation:**
- Widgets are stacked to the top, bottom, left, and right.
- The fill option allows the labels to expand horizontally or vertically.

### Example 05b: Layout Management Methods: grid()

Description:
- Organizes widgets in a table-like grid of rows and columns.
- Widgets are placed using `row` and `column` options.

Use Cases:
- Ideal for forms or tabular layouts.
- When precise alignment of widgets is required.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Grid Layout Example")
root.geometry("300x200")

# Add widgets using grid
tk.Label(root, text="Row 0, Col 0", bg="lightblue").grid(row=0, column=0)
tk.Label(root, text="Row 0, Col 1", bg="lightgreen").grid(row=0, column=1)
tk.Label(root, text="Row 1, Col 0", bg="pink").grid(row=1, column=0)
tk.Label(root, text="Row 1, Col 1", bg="yellow").grid(row=1, column=1)

# Add a widget that spans columns
tk.Label(root, text="Spanning Both Columns", bg="gray").grid(row=2, column=0, columnspan=2)

# Run the application
root.mainloop()
```

### Example 05c: Layout Management Methods: place()

Description:
- Positions widgets at exact coordinates using `x` and `y` values.
- Does not adjust automatically when the window resizes.

Use Cases:
- Perfect for custom layouts or pixel-perfect positioning.
- Useful when layout needs to match specific designs.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Place Layout Example")
root.geometry("300x250")

# Add widgets using place
tk.Label(root, text="Position (50, 50)", bg="lightblue").place(x=50, y=50)
tk.Label(root, text="Position (100, 100)", bg="lightgreen").place(x=100, y=100)
tk.Label(root, text="Position (150, 150)", bg="pink").place(x=150, y=150)
tk.Label(root, text="Position (50, 200)", bg="yellow").place(x=50, y=200)

# Run the application
root.mainloop()
```

**Explanation:**
- Widgets are placed at exact coordinates using the `x` and `y` options.
- This method is ideal for precise, static layouts but requires manual adjustments for different screen sizes.

## Example 06: Entry Widget Using Tkinter

The Entry widget in Tkinter is used to create a single-line input field where users can type text. It is typically used to gather user input such as names, passwords, or other small pieces of data.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Entry Widget Example")
root.geometry("300x150")

# Function to retrieve and display input
def display_input():
    user_input = entry.get()  # Get text from the entry widget
    label.config(text=f"Hello, {user_input}!")  # Update label with input

# Create an Entry widget
entry = tk.Entry(root, width=25)
entry.pack(pady=10)

# Create a Button to display the input
button = tk.Button(root, text="Submit", command=display_input)
button.pack(pady=10)

# Create a Label to show the result
label = tk.Label(root, text="Enter your name above", font=("Arial", 12))
label.pack(pady=10)

# Run the application
root.mainloop()
```

**Explanation:**
Entry Widget:
- The `Entry` widget lets the user type input.
- The `get()` method retrieves the text entered by the user.

Button:
- The `command=display_input` argument binds the button click to a function that reads the input and updates the label.

Label:
- Displays a message after the user clicks the button.

## Example 07: Listbox

The Listbox widget in Tkinter is used to display a list of items from which the user can select one or multiple items. It is a scrollable widget and is commonly used for tasks like item selection, displaying options, or presenting a list of data.

```python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title("Listbox Example")
root.geometry("300x200")

# Function to display the selected item
def show_selection():
    selected_item = listbox.get(listbox.curselection())  # Get the selected item
    label.config(text=f"Selected: {selected_item}")     # Update the label

# Create a Listbox widget
listbox = tk.Listbox(root, height=5)
listbox.pack(pady=10)

# Add items to the Listbox
items = ["Python", "Java", "C++", "JavaScript", "Ruby", "Matlab"]
for item in items:
    listbox.insert(tk.END, item)

# Create a Button to display the selection
button = tk.Button(root, text="Show Selection", command=show_selection)
button.pack(pady=5)

# Create a Label to show the result
label = tk.Label(root, text="Select an item from the list", font=("Arial", 12))
label.pack(pady=10)

# Run the application
root.mainloop()
```

**Explanation:**

1. Listbox Widget:
- Displays a list of items (items list).
- Users can select one item by clicking on it.

2. Button:
- Calls the show_selection() function, which:
- Retrieves the selected item using `listbox.curselection()` and `listbox.get()`.
- Updates the label with the selected item's text.

3. Label:
- Displays the selected item.

```python
import tkinter as tk
from tkinter import ttk
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Function to create a scatter plot
def plot_penguins():
    # Load the penguins dataset
    data = sns.load_dataset("penguins")
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(5, 4))
    sns.scatterplot(data=data, x="bill_length_mm", y="bill_depth_mm", hue="species", ax=ax)
    ax.set_title("Penguins: Bill Length vs. Bill Depth")
    ax.set_xlabel("Bill Length (mm)")
    ax.set_ylabel("Bill Depth (mm)")
    
    # Display the plot in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack()

# Create the main Tkinter window
root = tk.Tk()
root.title("Penguins Dataset Visualization")
root.geometry("600x500")

# Create a frame to hold the plot
frame = ttk.Frame(root)
frame.pack(pady=20)

# Add a button to plot the data
plot_button = ttk.Button(root, text="Visualize Penguins Data", command=plot_penguins)
plot_button.pack(pady=10)

# Run the application
root.mainloop()
```

## Multiple plot in a single GUI

Let's combine all of our Visualisation and GUI concepts together. The following demo creates a Tkinter-based GUI for visualising datasets such as Penguins, Iris, and Tips using Seaborn and Matplotlib. The application offers buttons for generating various plots like Scatter Plot, joint Plot, and Violin Plot. Each button triggers a function that loads the selected dataset and generates the respective visualisation. The plots are embedded within the GUI using `FigureCanvasTkAgg`. A frame dynamically updates the display area, clearing any previous plot. This GUI provides an interactive way to explore these datasets and understand trends, distributions, and patterns.

```python
import tkinter as tk
from tkinter import ttk
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np

# Function to plot the Multi-dimensional Scatter Plot
def plot_scatter():
    np.random.seed(42)
    x = np.random.rand(50) * 100  # X values
    y = np.random.rand(50) * 100  # Y values
    sizes = np.random.rand(50) * 1000  # Bubble sizes
    colors = np.random.rand(50)  # Color values

    fig, ax = plt.subplots(figsize=(8, 6))
    scatter = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
    
    ax.set_title("Multi-dimensional Scatter Plot", fontsize=14)
    ax.set_xlabel("X-axis: Feature 1", fontsize=12)
    ax.set_ylabel("Y-axis: Feature 2", fontsize=12)

    # Color bar for additional dimension
    color_bar = plt.colorbar(scatter, ax=ax)
    color_bar.set_label("Color Dimension")
    
    display_plot(fig)

# Function to plot the Joint Plot (Penguins dataset)
def plot_joint():
    penguins = sns.load_dataset("penguins")
    g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
    g.fig.suptitle("Joint Plot: Penguins Data", fontsize=14)
    g.fig.tight_layout()
    g.fig.subplots_adjust(top=0.95)  # Adjust title to fit
    display_plot(g.fig)

# Function to plot the Swarm Plot (Tips dataset)
def plot_swarm():
    tips = sns.load_dataset("tips")
    g = sns.catplot(
        data=tips, kind="swarm",
        x="time", y="total_bill", hue="sex", col="day",
        aspect=.5
    )
    g.fig.suptitle("Swarm Plot: Tips Data", fontsize=14)
    g.fig.tight_layout()
    g.fig.subplots_adjust(top=0.95)  # Adjust title to fit
    display_plot(g.fig)

# Function to plot the Violin Plot (Tips dataset) based on Day
def plot_violin_day():
    data = sns.load_dataset("tips")
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.violinplot(hue="day", y="total_bill", data=data, ax=ax, palette="bright")
    ax.set_title("Violin Plot: Tips Data", fontsize=14)
    display_plot(fig)

# Function to plot the Violin Plot (Tips dataset) based on Sex
def plot_violin_sex():
    data = sns.load_dataset("tips")
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.violinplot(hue="sex", x = 'day', y="total_bill", data=data, ax=ax, palette="bright")
    ax.set_title("Violin Plot: Tips Data", fontsize=14)
    display_plot(fig)

#Function to plot scatter histogram penguin plot
def histogram_plot_penguin():
    data = sns.load_dataset("penguins")
    g = sns.histplot(data, x="flipper_length_mm", kde=True, color = "green")
    g.set_title("Penguin Histogram", fontsize = 14)
    g.figure.tight_layout()
    g.figure.subplots_adjust(top = 0.95)
    display_plot(g.figure)
    
# Function to display plot in the Tkinter window
def display_plot(fig):
    for widget in frame_plot.winfo_children():
        widget.destroy()  # Clear previous plot
    canvas = FigureCanvasTkAgg(fig, master=frame_plot)
    canvas.draw()
    canvas.get_tk_widget().pack()

# Create the main Tkinter window
root = tk.Tk()
root.title("Data Visualization with Seaborn and Matplotlib")
root.geometry("1000x600")

# Create a frame to hold the buttons on the left
frame_buttons = ttk.Frame(root)
frame_buttons.pack(side="left", padx=20, pady=20)

# Create a frame to hold the plot area
frame_plot = ttk.Frame(root)
frame_plot.pack(side="right", padx=20, pady=20, fill="both", expand=True)

# Add buttons for different plots in the left frame
button_scatter = ttk.Button(frame_buttons, text="Scatter Plot", command=plot_scatter)
button_scatter.pack(pady=5, fill="x")

button_joint = ttk.Button(frame_buttons, text="Joint Plot", command=plot_joint)
button_joint.pack(pady=5, fill="x")

button_swarm = ttk.Button(frame_buttons, text="Swarm Plot", command=plot_swarm)
button_swarm.pack(pady=5, fill="x")

button_violin_sex = ttk.Button(frame_buttons, text="Violin Plot(Sex)", command=plot_violin_sex)
button_violin_sex.pack(pady=5, fill="x")

button_violin_day = ttk.Button(frame_buttons, text="Violin Plot(Day)", command=plot_violin_day)
button_violin_day.pack(pady=5, fill="x")

button_histogram = ttk.Button(frame_buttons, text = "Penguin Histogram", command=histogram_plot_penguin)
button_histogram.pack(pady=5, fill="x")

# Run the application
root.mainloop()
```

**Explanation of the Code:**

Tkinter GUI Setup:
- The root window is created, and a frame (frame) is added to display the plots.
- There are five buttons (Scatter Plot, Joint Plot, Swarm Plot, and two Violin Plots). Each button triggers a function that generates the corresponding plot.
  
Plot Functions:
- `plot_scatter()` generates a scatter plot with bubble sizes and color dimension using random data.
- `plot_joint()` uses the Penguins dataset to generate a joint plot of bill_length_mm vs. bill_depth_mm.
- `plot_swarm()` generates a swarm plot from the Tips dataset using total_bill vs. time, grouped by sex.
- `plot_violin_day()` and `plot_violin_sex()` generate a violin plots from the Tips dataset, showing the distribution of total_bill by day and sex.

Displaying the Plot:
- Each plot is displayed in the Tkinter window using FigureCanvasTkAgg. The previous plot is cleared before the new one is shown.

GUI Interaction:
- The user can click any of the buttons to view the corresponding plot.
- The plots are displayed directly in the window, except the joint plot which is slightly adjusted for better layout.
- This setup allows you to interactively view multiple types of visualizations from popular datasets directly within a Tkinter GUI.
