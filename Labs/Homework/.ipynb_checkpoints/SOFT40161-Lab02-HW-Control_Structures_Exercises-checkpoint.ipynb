### 14. Mini grid
**Task:** Print a 3x3 grid of numbers.
Use two loops for rows and columns.

**Expected Output:**
```
1 2 3
1 2 3
1 2 3
```

````python
# Write your code here

#Set a row with entries 1,2,3. Acces each index and output it in a row.

row = [[1, 2, 3], [1, 2, 3], [1, 2,3]]

for i in range(3):
    for j in range(3):
        entry = row[i][j]
        print(entry, end=" ")
    print()
````

### 16. Sum until limit
**Task:** Keep adding numbers from 1 upwards until the total exceeds 20.
Use a while loop with a stopping condition.

**Expected Output:**
```
Sum exceeded 20 at number 7!
```

````python
# Write your code here
count = 0
number = 1

while True:
    print(count)
    if count > 20:
        break
    count += number 
    number += 1

print("Sum exceeded 20 at number", number,"!")
````
